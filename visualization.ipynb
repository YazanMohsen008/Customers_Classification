# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset 

```python
dataset = pd.read_csv('customers.csv')
```

##  visualize gender distribution over segments.

```python
male   = {'A':0, 'B':0, 'C':0, 'D':0}
female = {'A':0, 'B':0, 'C':0, 'D':0}
for row in dataset.values:
    Segment = row[8]
    Gender = row[0]
    if Gender == 'Male':
        male[Segment] += 1
    else:
        female[Segment] += 1
# set width of bar
barWidth = 0.25

# set height of bar 
Male = [male['A'],male['B'],male['C'],male['D']]
Female = [female['A'],female['B'],female['C'],female['D']]
# Set position of bar on X axis
br1 =  [x + barWidth/2 for x in np.arange(4)]
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, Male, color ='blue', width = barWidth,edgecolor ='grey', label ='Male')
plt.bar(br2, Female, color ='pink', width = barWidth,edgecolor ='grey', label ='Female')

# Adding Xticks
plt.xlabel('Segments', fontweight ='bold', fontsize = 15)
plt.ylabel('n.of Customers', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(4)],  ['A', 'B', 'C', 'D'])

plt.legend()
plt.show()
```

##  visualize Marital status distribution over segments.

```python
married = {'A':0,'B':0,'C':0,  'D':0}
single= {'A':0,'B':0,'C':0, 'D':0}
for row in dataset.values:
    Segment = row[8]
    if row[1]=='Yes':
        married[Segment] += 1
    else:
        single[Segment] += 1
# set width of bar
barWidth = 0.25

# set height of bar 
Married = [married['A'],married['B'],married['C'],married['D']]
Single = [single['A'],single['B'],single['C'],single['D']]
# Set position of bar on X axis
br1 =  [x + barWidth/2 for x in np.arange(4)]
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, Married, color ='b', width = barWidth,edgecolor ='grey', label ='Married')
plt.bar(br2, Single, color ='r', width = barWidth,edgecolor ='grey', label ='Single')

# Adding Xticks
plt.xlabel('Segments', fontweight ='bold', fontsize = 15)
plt.ylabel('n.of Customers', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(4)],  ['A', 'B', 'C', 'D'])

plt.legend()
plt.show()
```

##  Visualize Age distribution over segments.

```python
# Calculate the average age
Average_Age = np.average(dataset.values[:, 2])

young   = {'A':0, 'B':0, 'C':0, 'D':0}
old = {'A':0, 'B':0, 'C':0, 'D':0}
for row in dataset.values:
    Segment = row[8]
    Age = row[2]
    if Age <= Average_Age:
        young[Segment] += 1
    else:
        old[Segment] += 1
# set width of bar
barWidth = 0.25

# set height of bar 
Young = [young['A'],young['B'],young['C'],young['D']]
Old = [old['A'],old['B'],old['C'],old['D']]
# Set position of bar on X axis
br1 =  [x + barWidth/2 for x in np.arange(4)]
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, Young, color ='red', width = barWidth,edgecolor ='grey', label ='Young')
plt.bar(br2, Old, color ='gray', width = barWidth,edgecolor ='grey', label ='Old')

# Adding Xticks
plt.xlabel('Segments', fontweight ='bold', fontsize = 15)
plt.ylabel('n.of Customers', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(4)],  ['A', 'B', 'C', 'D'])

plt.legend()
plt.show()
```
